```python
import pandas as pd

df = pd.read_csv('kpc_hai_results.tsv', sep='\t', dtype={'hsn': object})
```

```python
plasmids = df[['hsn', 'plasmidfinder_plasmids']]
amr_genes = df[['hsn', 'amrfinderplus_amr_genes']]
```

```python
# amr_gene = 'blaKPC'
print('hi')
def add_amr_col(row):
    amr_genes = row['amrfinderplus_amr_genes']
    try:
       variant = amr_genes[amr_genes.index(amr_gene):amr_genes.index(amr_gene)+len(amr_gene)+2]
    except (ValueError, AttributeError):
        variant = 'blaKPC not found'
    return variant

amr_genes['kpc'] = amr_genes.apply(add_amr_col, axis=1)
amr_genes
```

```python
plasmids = plasmids.dropna()
plasmids
```

```python
plasmid_explode = plasmids.set_index('hsn').apply(lambda x: x.str.split(',').explode())
plasmid_explode
```

```python
merged = pd.merge(plasmid_explode, amr_genes, on='hsn', how='left')
merged
```

```python
pd.concat(g for _, g in merged.groupby(['kpc', 'plasmidfinder_plasmids']) if len(g) > 1)
```
